```python
import re
all_prods = products_container.find_all('div', class_='col')
for i, prod in enumerate(all_prods):
    price = prod.find('span', class_=re.compile('item__product__price'))
    if not price:
        continue
    price = price['data-price']
    name = ' '.join(prod.find('span', class_='line-clamp--2').text.split())
    try:
        units = prod.find('span', class_=re.compile('tile-item__product__size'))
        unit_type = units.abbr['title']
        quantity = units.next.strip()
    except TypeError as e:
        unit_type, quantity = None, None
    print(i, price, name, unit_type, quantity)
```
